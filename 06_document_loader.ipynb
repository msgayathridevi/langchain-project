# Structured File Loader

from langchain_community.document_loaders import UnstructuredFileLoader, JSONLoader
from pathlib import Path

files = Path("data").glob("*")

all_docs = []

for file in files:
  if file.suffix == ".json":
    un_loader = JSONLoader(file, jq_schema=".", text_content=False)
  else:
    un_loader = UnstructuredFileLoader(file)
  all_docs.extend(un_loader.load())

print(len(all_docs))
print(all_docs[0].page_content)
